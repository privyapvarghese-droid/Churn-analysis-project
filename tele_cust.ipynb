```python
numerical_columns = tele.select_dtypes(include=np.number).columns

for column in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(tele[column], kde=True, bins=20)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
tele['customerID']=le.fit_transform(tele['customerID'])
tele['gender']=le.fit_transform(tele['gender'])
tele['Partner']=le.fit_transform(tele['Partner'])
tele['Dependents']=le.fit_transform(tele['Dependents'])
tele['PhoneService']=le.fit_transform(tele['PhoneService'])
tele['MultipleLines']=le.fit_transform(tele['MultipleLines'])
tele['InternetService']=le.fit_transform(tele['InternetService'])
tele['OnlineSecurity']=le.fit_transform(tele['OnlineSecurity'])
tele['OnlineBackup']=le.fit_transform(tele['OnlineBackup'])
tele['DeviceProtection']=le.fit_transform(tele['DeviceProtection'])
tele['TechSupport']=le.fit_transform(tele['TechSupport'])
tele['StreamingTV']=le.fit_transform(tele['StreamingTV'])
tele['StreamingMovies']=le.fit_transform(tele['StreamingMovies'])
tele['Contract']=le.fit_transform(tele['Contract'])
tele['PaperlessBilling']=le.fit_transform(tele['PaperlessBilling'])
tele['PaymentMethod']=le.fit_transform(tele['PaymentMethod'])
tele['Churn']=le.fit_transform(tele['Churn'])
tele
```

```python
churn_rate = tele['Churn'].value_counts(normalize=True)
sns.barplot(x=churn_rate.index, y=churn_rate.values, palette='Set2')
plt.title('Churn Rate')
plt.ylabel('Proportion')
plt.xlabel('Churn')
plt.show()
```

```python
sns.countplot(x='gender', hue='Churn', data=tele, palette='Set1')
plt.title('Churn by Gender')
plt.show()
```

Gender does not make any significant difference in churn count

```python
sns.countplot(x='SeniorCitizen', hue='Churn', data=tele, palette='Set2')
plt.title('Churn by Senior Citizen Status')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

The number of senior citizens who churned is lower compared to others.

```python
sns.countplot(x='Partner', hue='Churn', data=tele, palette='Set3')
plt.title('Churn by Partner')
plt.show()
```

The persons without a partner is more likely to leave.

```python
sns.countplot(x='InternetService', hue='Churn', data=tele, palette='Set1')
plt.title('Churn by Internet Service')
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=tele, palette='Set2')
plt.title('Churn by Contract Type')
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=tele, palette='Set3')
plt.title('Tenure by Churn')
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=tele, palette='Set1')
plt.title('Monthly Charges by Churn')
plt.show()
```

```python
sns.boxplot(x='Churn', y='TotalCharges', data=tele, palette='Set2')
plt.title('Total Charges by Churn')
plt.show()
```

```python
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=tele, palette='Set1', alpha=0.7)
plt.title('Tenure vs Monthly Charges by Churn')
plt.show()
```

```python
tele['Churn_numeric'] =tele['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

sns.heatmap(tele[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn_numeric']].corr(),
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation with Churn')
plt.show()
```

```python
tele['tenure_group'] = pd.cut(tele['tenure'], bins=[0, 12, 24, 48, 60, 72],
                            labels=['0-12', '13-24', '25-48', '49-60', '61-72'])

sns.countplot(x='tenure_group', hue='Churn', data=tele, palette='Set2')
plt.title('Churn by Tenure Group')
plt.show()
```

```python
# Drop the 'tenure_group' column as it's not numeric and causes errors in correlation calculation
tele_numeric = tele.drop('tenure_group', axis=1)
plt.figure(figsize=(15,10))
sns.heatmap(tele_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
